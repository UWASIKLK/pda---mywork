## Introduction
***

The Palmer Penguins dataset contains various measurements of three different penguin species (Adelie, Gentoo and Chinstrap) and it commonly used for data analysis as an alternative to the Iris dataset.

These data were collected from 2007 - 2009 by [Dr. Kristen Gorman](https://www.uaf.edu/cfos/people/faculty/detail/kristen-gorman.php) 
with the [Palmer Station](https://pallter.marine.rutgers.edu/), a member of the [Long Term Ecological Research Network](https://lternet.edu/) on three islands. 

## Investigation
***


The dataset has details of 344 penguins from three species of penguin: _Adélie (152 penguins)_, _Chinstrap (68 penguins)_ and _Gentoo (124 penguins)_.

The dataset contains following variables:
 - **species:**&nbsp; penguins species (_Adélie_, _Chinstrap_ or _Gentoo_)
 - **island:**&nbsp; island name (_Dream_, _Torgersen_ or _Biscoe_)
 - **bill_length_mm:**&nbsp; bill (culmen) length in millimetres
 - **bill_depth_mm:**&nbsp; depth length in millimetres
 - **flipper_length_mm:**&nbsp; flipper length in millimetres
- **body_mass_g:**&nbsp; weights in grams
- **sex:**&nbsp; the gender of penguin

## Data clean up

***

After running the info() function I can see that some information related to bill length & depth, flipper length, body mass and sex is missing. I will investigate this further using the isnull() function which will show rows that contains null values. the rows which will return boolean **'True'** will have a null values.

```python
# To find out if there are any missing values

penguins.isnull()
```

After running the above function, I can see that two rows have missing values. The one is for Adelie penguin and anothere for Gentoo penguin. In case of large dataset, we have an option to delete the missing rows, as they will not have a significant impact on the overall result of analysis. Or we can try to replace the missing values. I decided to go for second option and replace the missing values.

```python
# This will show the details of two rows with missing values

penguins.iloc[[3,339],0:7]
```

Now I want to find out the average values of the missing columns. First, I find out the details of Adelie's penguin from Torgersen Island. I want to see the numbers of male & female related to Adelie, because this also needs to be filled.

```python
# Showing details related to Adelie, Torgersen island

a_penguins = penguins.loc[(penguins.species == 'Adelie') & (penguins.island == 'Torgersen')]
a_penguins.describe()
```

```python
# And the gender count (how many of female and male)

a_penguins['sex'].value_counts()
```

I will use the **'mean'** values for the missing Adelie row. Since there is not much difference in the numbers of male and female, I opted for a female in this case.

```python
# Filling in the missing data for Adelie row (values are on two decimal places)

penguins.iloc[3,2:7] = 38.95, 18.42, 191.19, 3706.37, 'FEMALE'

```

```python
# checking if missing data were replaced by mean values for Adelie row

penguins.iloc[[3,339],0:7]
```

The missing data for Adelie row are now filled in. I will repeat the same steps for Gentoo penguin row.

```python
# Showing details related to Gentoo, Biscoe island

g_penguins = penguins.loc[(penguins.species == 'Gentoo') & (penguins.island == 'Biscoe')]
g_penguins.describe()
```

```python
# And the gender count (how many of female and male)

g_penguins['sex'].value_counts()
```

I will use the mean values for missing Gentoo row and I opted for a male this time.

```python
# Filling in the missing data for Gentoo row (values are on two decimal places)

penguins.iloc[339,2:7] = 47.50, 14.98, 217.18, 5076.01, 'MALE'
```

```python
# Checking if missing data were replaced by mean values for Adelie row

penguins.iloc[[3,339],0:7]
```

I want to check how many rows we have, where the **'sex'** column has no value.

```python
# This will show how many rows have missing value in sex column.

penguins[penguins['sex'].isna()]
```

In general, the body mass for male penguin is higher than in female penguin. I want to find out what is the average body weight of female Adelie penguins from the island of Torgersen. I will repeat the same steps for Gentoo penguins.

```python
# finding the average body mass

aa_penguins = penguins.loc[(penguins.species == 'Adelie') & (penguins.island == 'Torgersen') & (penguins.sex == 'FEMALE')]
aa_penguins.describe()
```

According to the above result, the average body weight of Adelie penguins from Torgersen Island is 3408.25 grams. Therefore, for row 10 I fill in 'FEMALE' and for rows 8,9 & 11 I fill in 'MALE'.

```python
# Filling in missing gender data.

penguins.iloc[10,6] = 'FEMALE'
penguins.iloc[8,6] = 'MALE'
penguins.iloc[9,6] = 'MALE'
penguins.iloc[11,6] = 'MALE'
```

I assume, that the row 47, Adelie from Dream Island with the body mass 2975.0 grams, is 'FEMALE'. It is clear from the dataset that males are larger/heavier than females.

```python
# Filling in missing gender. 

penguins.iloc[47,6] = 'FEMALE'
```

I will repeat the steps to find out the average body weight of Gentoo penguins from Biscoe Island.

```python
gg_penguins = penguins.loc[(penguins.species == 'Gentoo') & (penguins.island == 'Biscoe') & (penguins.sex == 'FEMALE')]
gg_penguins.describe()
```

The average body mass as per result above is 4679.74. If the body mass is below this figure on missing rows I fill in 'FEMALE', otherwise it will be 'MALE'.

```python
# Filling in missing data.

penguins.iloc[246,6] = 'FEMALE'
penguins.iloc[286,6] = 'FEMALE'
penguins.iloc[324,6] = 'MALE'
penguins.iloc[336,6] = 'MALE'
```
