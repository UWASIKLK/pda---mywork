## Introduction
***

The Palmer Penguins dataset contains various measurements of three different penguin species (Adelie, Gentoo and Chinstrap) and it is commonly used for data analysis as an alternative to the Iris dataset.

These data were collected from 2007 - 2009 by [Dr. Kristen Gorman](https://www.uaf.edu/cfos/people/faculty/detail/kristen-gorman.php) 
with the [Palmer Station](https://pallter.marine.rutgers.edu/), a member of the [Long Term Ecological Research Network](https://lternet.edu/) on three islands. 

## Investigation
***



The dataset has details of 344 penguins from three species of penguin: _Adélie (152 penguins)_, _Chinstrap (68 penguins)_ and _Gentoo (124 penguins)_.

The dataset contains following variables:
 - **species:**&nbsp; penguins species (_Adélie_, _Chinstrap_ or _Gentoo_)
 - **island:**&nbsp; island name (_Dream_, _Torgersen_ or _Biscoe_)
 - **bill_length_mm:**&nbsp; bill (culmen) length in millimetres
 - **bill_depth_mm:**&nbsp; depth length in millimetres
 - **flipper_length_mm:**&nbsp; flipper length in millimetres
- **body_mass_g:**&nbsp; weights in grams
- **sex:**&nbsp; the gender of penguin

### Data clean up

***

After running the info() function I can see that some information related to bill length & depth, flipper length, body mass and sex is missing. I will investigate this further using the isnull() function which will show rows that contains null values. The rows which will return boolean **'True'** will have a null values.

```python
# To find out if there are any missing values

penguins.isnull()
```

I can see that two rows have missing values. The one is for Adelie penguin and anothere for Gentoo penguin. In case of large dataset, there is an option to delete the missing rows, as they will not have a significant impact on the overall result of analysis. Or, we can try to replace the missing values. I decided to go for the second option and replace the missing values.

```python
# This will show the details of two rows with missing values.

penguins.iloc[[3,339],0:7]
```

Now, I want to find out the average values of the missing columns. First, I will find out the details of Adelie's penguin from Torgersen Island.

```python
# Showing details related to Adelie, Torgersen Island

a_penguins = penguins.loc[(penguins.species == 'Adelie') & (penguins.island == 'Torgersen')]
a_penguins.describe()
```

```python
# And the gender count (how many of female and male are there)

a_penguins['sex'].value_counts()
```

I will use the **'mean'** values for the missing Adelie row. Since there is not much difference in the numbers of male and female, I opted for a female in this case.

```python
# Filling in the missing data for Adelie row (values are on two decimal places)

penguins.iloc[3,2:7] = 38.95, 18.42, 191.19, 3706.37, 'FEMALE'

```

```python
# Checking if missing data were replaced by mean values for Adelie row

penguins.iloc[[3,339],0:7]
```

The missing data for Adelie row are now filled in. I will repeat the same steps for Gentoo penguin row.

```python
# Showing details related to Gentoo, Biscoe Island

g_penguins = penguins.loc[(penguins.species == 'Gentoo') & (penguins.island == 'Biscoe')]
g_penguins.describe()
```

```python
# And the gender count (how many of female and male are there)

g_penguins['sex'].value_counts()
```

I will use the mean values for missing Gentoo row and I opted for a male this time.

```python
# Filling in the missing data for Gentoo row (values are on two decimal places)

penguins.iloc[339,2:7] = 47.50, 14.98, 217.18, 5076.01, 'MALE'
```

```python
# Checking if missing data were replaced by mean values for Gentoo row

penguins.iloc[[3,339],0:7]
```

I want to check how many rows we have, where the **'sex'** column has no value.

```python
# This will show how many rows have missing value in sex column.

penguins[penguins['sex'].isna()]
```

In general, the body mass for male penguin is higher than in female penguin. I want to find out what is the average body weight of female Adelie penguins from the island of Torgersen. I will repeat the same steps for Gentoo penguins.

```python
# Finding the average body mass for Adelie from Torgersen Island

aa_penguins = penguins.loc[(penguins.species == 'Adelie') & (penguins.island == 'Torgersen') & (penguins.sex == 'FEMALE')]
aa_penguins.describe()
```

According to the above result, the average body weight of Adelie penguins from Torgersen Island is 3408.25 grams. Therefore, for row 10 I fill in 'FEMALE' and for rows 8,9 & 11 I fill in 'MALE'.

```python
# Filling in missing gender data.

penguins.iloc[10,6] = 'FEMALE'
penguins.iloc[8,6] = 'MALE'
penguins.iloc[9,6] = 'MALE'
penguins.iloc[11,6] = 'MALE'
```

I assume, that the row 47, Adelie from Dream Island with the body mass 2975.0 grams, is 'FEMALE'. It is clear from the dataset that males are larger/heavier than females.

```python
# Filling in missing gender. 

penguins.iloc[47,6] = 'FEMALE'
```

I will repeat the steps to find out the average body weight of Gentoo penguins from Biscoe Island.

```python
gg_penguins = penguins.loc[(penguins.species == 'Gentoo') & (penguins.island == 'Biscoe') & (penguins.sex == 'FEMALE')]
gg_penguins.describe()
```

The average body mass as per result above is 4679.74 grams. If the body mass is below this figure on missing rows I fill in 'FEMALE', otherwise it will be 'MALE'.

```python
# Filling in missing data.

penguins.iloc[246,6] = 'FEMALE'
penguins.iloc[286,6] = 'FEMALE'
penguins.iloc[324,6] = 'MALE'
penguins.iloc[336,6] = 'MALE'
```

```python
# Checking for null values on dataset

penguins.isnull().sum()
```

The above result indicates that there are no null/missing values in dataset. 

## Visualization
***

### Finding a mean


```python
# I want to find out the mean for island, species and sex for each variables.

variables = ['bill_length_mm', 'bill_depth_mm','flipper_length_mm', 'body_mass_g']
mean_result = penguins.groupby(['island', 'species', 'sex'])[variables].mean()
print(mean_result)
```

```python
# This will show the mean for the weight by gender, species and island.

penguins.groupby(['sex','species', 'island'])['body_mass_g'].mean()
```

***
### Bar chart


```python
# I will create a bar chart based on the island, gender, species and average of the weight.

penguins.groupby(['island','sex','species'])['body_mass_g'].mean().unstack().plot(kind ='bar', color =['indianred','steelblue','forestgreen'])
plt.xlabel('Island & Gender', color ='darkslategray')
plt.ylabel('Average weight in grams', color='darkslategray')
plt.title('Body mass by Island & Species', color ='darkslategray')
plt.show()
```

Adelie penguins are on all 3 islands, but Gentoo penguins are on Biscoe Island only and Chinstrap on Torgersen Island.
There is a significant difference in weight between the Gentoo penguins and other two, Adelie & Chinstrap.
I can see that Gentoo are much heavier than others.

```python
# Bar chart by bill lenght, species and gender.

penguins.groupby(['sex','species'])['bill_length_mm'].mean().unstack().plot(kind ='bar', color =['indianred','steelblue', 'forestgreen'])
plt.xlabel('Gender', color ='darkslategray')
plt.ylabel('Bill Length in millimeters', color ='darkslategray')
plt.title('Bill Lenght by Species', color ='darkslategray')
plt.show()
```

The bill length for Adelie penguins is the shortest compared to the Chinstrap and Gentoo penguins.
And obviously, the female has it shortest than male penguins.

```python
# Bar chart by bill depth, species and gender.

penguins.groupby(['sex','species'])['bill_depth_mm'].mean().unstack().plot(kind ='bar', color =['indianred','steelblue', 'forestgreen'])
plt.xlabel('Gender', color ='darkslategray')
plt.ylabel('Bill Depth in millimeters', color ='darkslategray')
plt.title('Bill Depth by Species', color ='darkslategray')
plt.show()
```

As per above chart, Gentoo penguis have the shortest bill depth. Adelie and Chinstrap penguins do not differ as much.

```python
# Bar chart by flipper lenght, species and gender.
penguins.groupby(['sex','species'])['flipper_length_mm'].mean().unstack().plot(kind ='bar', color =['indianred','steelblue', 'forestgreen'])
plt.xlabel('Gender', color ='darkslategray')
plt.ylabel('Flipper Length in millimeters', color ='darkslategray')
plt.title('Flipper Length by Species', color ='darkslategray')
plt.show()
```

Comparing the flipper length, the Gentoo penguins have the longest one. Adelie penguins has the shortest flipper length.

***
### Box plot


```python
# Creating a box plot using the island, weight and it is sort out by 3 species.

box = sns.boxplot(x = 'island', y = 'body_mass_g', hue ='species', data = penguins, palette='Set2')
plt.xlabel('Island', color ='darkslategray')
plt.ylabel('Weight in grams', color ='darkslategray')
plt.title('Body mass by Island & Species', color ='darkslategray')
plt.show()
```

What I can see here is, that the Adelie penguin - presenting on all 3 Islands, and Chinstrap penguins - from Dream Island, are almost on the same 
level in terms of the weight / body mass. Whereas, the body mass of Gentoo penguins is significantly higher.


```python
# Creating a box plot using the island, bill length and it is sort out by 3 species.

box = sns.boxplot(x = 'island', y = 'bill_length_mm', hue ='species', data = penguins, palette='Set2')
plt.xlabel('Island', color ='darkslategray')
plt.ylabel('Bill Length in millimetres', color ='darkslategray')
plt.title('Bill Length by Island & Species', color ='darkslategray')
plt.show()
```

Adelie penguins based on all 3 islands have almost the same bill lenght. The Chinstrap and Gentoo penguins has much longer bill lenght.

```python
# Creating a box plot using the island, bill depth and it is sort out by 3 species.

box = sns.boxplot(x = 'island', y = 'bill_depth_mm', hue ='species', data = penguins, palette='Set2')
plt.xlabel('Island', color ='darkslategray')
plt.ylabel('Bill Depth in millimetres', color ='darkslategray')
plt.title('Bill Depth by Island & Species', color ='darkslategray')
plt.show()
```

Gentoo penguins have the shortest bill depth, while Adelie & Chinstrap has almost the same bill depth size.

```python
# Creating a box plot using the island, bill depth and it is sort out by 3 species.

box = sns.boxplot(x = 'island', y = 'flipper_length_mm', hue ='species', data = penguins, palette='Set2')
plt.xlabel('Island', color ='darkslategray')
plt.ylabel('Flipper Length in millimetres', color ='darkslategray')
plt.title('Flipper Length by Island & Species', color ='darkslategray')
plt.show()
```

The flipper lenght for Gentoo penguins is the longest. The second longer is Chinstrap penguins and the shortest are Adelie penguins.

***
### Scatter plot

```python
# This is a scatter plot chart using all information from dataset sort out by species.

sns.scatterplot(x='flipper_length_mm', y='body_mass_g', data=penguins, hue='species', palette='Set2')
plt.xlabel('Flipper Length in millimeters', color='darkslategray')
plt.ylabel('Body Mass in grams', color='darkslategray')
plt.title('Flipper Length vs Weight', color='darkslategray')
plt.show()
```

The above scatter plot is showing the relation between the weight and flipper length for each sepcies. I see there is a correlation between these variables, and heavier are the Gentoo penguins with the longest flipper length.

```python
# I created the scatter plot with the trend lines for all data points for each species so I could see the correlation between the species.
# I addedd the trend line for all data points combined to show the general overview of dataset for specific variables.

plt.figure(figsize=(15,10))
sns.scatterplot(x='body_mass_g', y='flipper_length_mm', data=penguins, hue='species')

# Add trend lines for all data points

sns.regplot(x='body_mass_g', y='flipper_length_mm', data=penguins, scatter=False, lowess=True, color='yellow')

# Separating the trend lines for each species.
species_list = penguins['species'].unique()
for species in species_list:
    block = penguins[penguins['species'] == species]
    sns.regplot(x='body_mass_g', y="flipper_length_mm", data=block, scatter=False, logx=True, label=species)
plt.title("Weigth vs Flipper length", color ='darkslategray')
plt.xlabel('Flipper Length in millimeters', color='darkslategray')
plt.ylabel('Body Mass in grams', color='darkslategray')
plt.legend()
plt.show()

```

The scattering graph above shows each species and the relationship between flipper length and weight. The trend lines for each species show an overall view, and the yellow trend line shows correlation between all 3 species.

***
### Pair plot

```python
# Below is a pair plot of each variables from dataset. 
plt.figure(figsize=(15,10))
sns.pairplot(data = penguins, hue = 'species', palette = 'Set2')
plt.show()
```

### KDE plot

```python
# I created the KDE graph which shows the probability density of flipper length, body mass within all 3 species as an example.

sns.kdeplot(x='body_mass_g', y='flipper_length_mm', data=penguins, hue='species', palette='Set2')
plt.xlabel('Body Mass in grams', color='darkslategray')
plt.ylabel('Flipper Length in millimeters', color='darkslategray')
plt.title('Flipper Length vs Weight', color='darkslategray')
plt.show()
```

Gentoo penguins have longer flipper and they are much heavier comparing to others, Adelie & Chinstrap.
